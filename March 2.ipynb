```python
x = np.linspace(0,10,200)
y = np.sin(x)
```

```python
plt.figure(figsize=(6,2))
plt.plot(x,y,'--c')
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("This is my title")
plt.grid()
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)

colours = np.random.rand(50)
sizes = 500*np.random.rand(50)
plt.scatter(x,y, c= '#A64D13', s = sizes, alpha=0.5, marker='x')
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Random Data")
```
